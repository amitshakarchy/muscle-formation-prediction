```python
def get_explainer(path):
    clf = joblib.load(path + "/clf.joblib")
    if isinstance(clf, CalibratedClassifierCV):
        clf = clf.base_estimator
    x_test = pd.read_csv(path + "/" + "X_test", encoding="cp1252")
    drop_cols = ["Spot track ID"]
    if "Unnamed: 0" in x_test.columns:
        drop_cols.append("Unnamed: 0")
    
    # plot feature_importance
    (pd.Series(clf.feature_importances_, index=x_test.drop(columns=drop_cols).columns).nlargest(20).plot(kind='barh'))   
    plt.title("rf feature_importances_")
    plt.show()
    
    explainer = shap.explainers.Tree(clf)
    shap_values = explainer.shap_values(x_test.drop(columns=drop_cols))

    return x_test, explainer, shap_values
```

```python
path_mot = consts.motility_model_path
mot_x_test, mot_explainer, mot_shap_values =get_explainer(path_mot)
shap.summary_plot(mot_shap_values[0], mot_x_test.drop(columns=["Spot track ID"]),  plot_type='violin')
```

```python
path_int = consts.intensity_model_path
int_x_test, int_explainer, int_shap_values =get_explainer(path_int)
shap.summary_plot(int_shap_values[0], int_x_test.drop(columns=["Spot track ID"]),  plot_type='violin')
```
